```python
from PIL import Image
import binascii

def hex_to_image(hex_string, width, height, output_path):
    # Convertir la chaîne hexadécimale en données binaires
    binary_data = binascii.unhexlify(hex_string)
    
    # Créer une image à partir des données binaires
    image = Image.frombytes('RGB', (width, height), binary_data)
    
    # Sauvegarder l'image au format PNG
    image.save(output_path)

# Exemple d'utilisation
hex_string = 'FF0000FF00FF0000FF00FF00FF00FFFFFF'  # Exemple de chaîne hexadécimale
width = 2   # Largeur de l'image
height = 2  # Hauteur de l'image
output_path = 'output_image.png'  # Chemin de sauvegarde de l'image

hex_to_image(hex_string, width, height, output_path)
```

```python
from PIL import Image
import binascii
import io

def hex_to_png(hex_string, output_path):
    # Convertir la chaîne hexadécimale en données binaires
    binary_data = binascii.unhexlify(hex_string)
    
    # Créer un objet de type fichier à partir des données binaires
    image_file = io.BytesIO(binary_data)
    
    # Ouvrir l'image avec Pillow
    image = Image.open(image_file)
    
    # Sauvegarder l'image au format PNG
    image.save(output_path, format='PNG')
```
