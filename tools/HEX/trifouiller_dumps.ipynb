```python
def get_t(x):
    return(x[36:44])
```

```python
def hex_to_date_timestamp_frmt(timestamp_hex):
    timestamp_hex = timestamp_hex.replace(' ','')
    timestamp_bytes = bytearray.fromhex(timestamp_hex)
    timestamp_dec = int.from_bytes(timestamp_bytes, byteorder = 'little')
    timestamp_datetime = datetime.datetime.fromtimestamp(timestamp_dec)
    return(timestamp_datetime.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
df['raw'].apply(get_t).apply(hex_to_date_timestamp_frmt)
```

```python
def timestamp_to_hex(timestamp: int) -> str:
    """Convertit un timestamp en une chaÃ®ne hexadÃ©cimale sur 8 octets."""
    return timestamp.to_bytes(length =4,  byteorder='little').hex() # 16 caractÃ¨res = 8 octets en hexadÃ©cimal

# Exemple d'utilisation
timestamp = 1741602642
hex_value = timestamp_to_hex(timestamp)
print(hex_value)  # Affichera la reprÃ©sentation hexadÃ©cimale
```

```python
def crc16_x25(data_str: str) -> str:
    data = data_str.encode('utf-8')  # ou 'ascii' selon le contexte
    crc = 0xFFFF
    for byte in data:
        crc ^= byte
        for _ in range(8):
            if crc & 0x0001:
                crc = (crc >> 1) ^ 0x8408
            else:
                crc >>= 1
    crc ^= 0xFFFF
    return f"{crc:04X}"  # format hexa sur 4 chiffres, majuscules
message = "Hello AX25"
crc_str = crc16_x25(message)
print(f"CRC HEX: {crc_str}")  # Affiche par ex. '6B1A'
```

```python
l3 = []
for i in l2:
    data = i[:-4]
    l3.append(data + str(crc16_x25(data)))
```
