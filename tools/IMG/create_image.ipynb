```python
from PIL import Image, ImageDraw, ImageFont

# Taille de l'image
largeur, hauteur = 400, 300

# Couleur de fond de l'image (blanc)
couleur_fond = (255, 255, 255)

# Créer une nouvelle image
image = Image.new('RGB', (largeur, hauteur), couleur_fond)

# Créer un objet pour dessiner sur l'image
dessin = ImageDraw.Draw(image)

# Charger une police (ou utiliser la police par défaut)
try:
    # Charger une police système (arial par exemple)
    police = ImageFont.truetype("arial.ttf", 40)
except IOError:
    # Si la police n'est pas trouvée, utiliser la police par défaut
    police = ImageFont.load_default()

# Titre
titre = "Mon Imageaqaaaaaaaaaaaaaaaaa"  # Peu importe la taille du titre

# Fonction pour ajuster dynamiquement la taille de la police
def ajuster_taille_police(texte, largeur_image, font, max_font_size=100):
    """Ajuste la taille de la police pour que le texte tienne dans la largeur donnée."""
    taille_police = max_font_size
    police_redimensionnee = ImageFont.truetype(font.path, taille_police)
    
    # Réduire la taille de la police jusqu'à ce que le texte tienne dans la largeur
    while dessin.textbbox((0, 0), texte, font=police_redimensionnee)[2] > largeur_image:
        taille_police -= 1
        police_redimensionnee = ImageFont.truetype(font.path, taille_police)
    
    return police_redimensionnee

# Ajuster la taille de la police pour que le titre prenne toute la largeur de l'image
police_titre_ajustee = ajuster_taille_police(titre, largeur - 40, police)  # En laissant une petite marge de 40px

# Position pour centrer le texte
position_titre = ((largeur - dessin.textbbox((0, 0), titre, font=police_titre_ajustee)[2]) // 2, 20)

# Couleur du texte (noir)
couleur_texte = (0, 0, 0)

# Ajouter le titre ajusté sur l'image
dessin.text(position_titre, titre, font=police_titre_ajustee, fill=couleur_texte)

# Sauvegarder l'image dans un fichier
image.save("image_avec_titre_ajuste.png")

# Afficher l'image dans une fenêtre (optionnel)
image.show()
```

```python
from PIL import Image, ImageDraw, ImageFont

# Taille de l'image
largeur, hauteur = 800, 800

# Couleur de fond de l'image (blanc)
couleur_fond = (255, 255, 255)

# Créer une nouvelle image
image = Image.new('RGB', (largeur, hauteur), couleur_fond)

# Créer un objet pour dessiner sur l'image
dessin = ImageDraw.Draw(image)

# Charger une police (ou utiliser la police par défaut)
try:
    # Charger une police système (arial par exemple)
    police = ImageFont.truetype("arial.ttf", 120)
except IOError:
    # Si la police n'est pas trouvée, utiliser la police par défaut
    police = ImageFont.load_default()

# Titre et description
titre = "Mon Image"
description = ("Ceci est une description qui doit être divisée en trois segments pour s'ajuster "
               "à l'espace disponible sous le titre, sans couper les mots et en maximisant "
               "la taille de la police pour bien occuper l'espace.")

# Fonction pour diviser la description en 3 segments à peu près égaux
def diviser_description(description, n_segments=3):
    """Divise la description en n_segments sans couper les mots."""
    mots = description.split(' ')
    longueur_segment = len(mots) // n_segments + 2
    segments = []
    
    # Diviser la description en segments
    for i in range(n_segments):
        if i < n_segments - 1:
            segments.append(' '.join(mots[i * longueur_segment:(i + 1) * longueur_segment]))
        else:
            segments.append(' '.join(mots[i * longueur_segment:]))  # Dernier segment prend le reste des mots
    
    return segments

# Diviser la description en trois segments
segments_description = diviser_description(description, 10)

# Ajuster la taille de la police pour les segments avec une marge de 80px
marge = 40
hauteur_disponible = hauteur - 250  # Hauteur disponible sous le titre (ajustée pour marges)
largeur_disponible = largeur - marge  # Largeur disponible pour le texte

# Fonction pour ajuster dynamiquement la taille de la police pour occuper la hauteur ET la largeur disponible
def ajuster_taille_police_segments(segments, largeur_image, hauteur_disponible, font, max_font_size=100):
    """Ajuste la taille de la police pour que les segments tiennent dans la largeur et la hauteur disponibles."""
    taille_police = max_font_size
    police_redimensionnee = ImageFont.truetype(font.path, taille_police)
    
    # Réduire la taille de la police jusqu'à ce que tout le texte tienne à la fois en largeur et en hauteur
    while True:
        hauteur_totale = 0
        largeur_max_segment = 0
        
        for segment in segments:
            bbox = dessin.textbbox((0, 0), segment, font=police_redimensionnee)
            hauteur_totale += bbox[3]
            largeur_max_segment = max(largeur_max_segment, bbox[2])  # Prendre la largeur max d'un segment
        
        # Si tout tient dans la largeur et la hauteur, on arrête
        if hauteur_totale <= hauteur_disponible and largeur_max_segment <= largeur_image:
            break
        
        # Sinon, on réduit la taille de la police
        taille_police -= 1
        police_redimensionnee = ImageFont.truetype(font.path, taille_police)
    
    return police_redimensionnee

# Ajuster la taille de la police pour que tous les segments tiennent dans la hauteur et la largeur disponible
police_ajustee = ajuster_taille_police_segments(segments_description, largeur_disponible, hauteur_disponible, police)

# Positionner le titre
position_titre = ((largeur - dessin.textbbox((0, 0), titre, font=police)[2]) // 2, 20)
dessin.text(position_titre, titre, font=police, fill=(0, 0, 0))

# Positionner les trois segments de la description
hauteur_titre = dessin.textbbox((0, 0), titre, font=police)[3]  # Hauteur du titre
position_segment_y = hauteur_titre + 40  # Commencer en dessous du titre

for segment in segments_description:
    # Centrer chaque segment horizontalement
    position_segment_x = (largeur - dessin.textbbox((0, 0), segment, font=police_ajustee)[2]) // 2
    dessin.text((position_segment_x, position_segment_y), segment, font=police_ajustee, fill=(0, 0, 0))
    
    # Avancer la position verticale pour le prochain segment
    position_segment_y += dessin.textbbox((0, 0), segment, font=police_ajustee)[3] + 10

# Sauvegarder l'image dans un fichier
image.save("image_avec_segments_description_corrige.png")

# Afficher l'image dans une fenêtre (optionnel)
image.show()
```

```python
# Dictionnaire pour stocker les résultats
dictionnaire_resultat = {}

# Chemin du fichier à lire
fichier_texte = "test.txt"

# Ouvrir le fichier en mode lecture
with open(fichier_texte, 'r', encoding='utf-8') as fichier:
    # Lire le fichier ligne par ligne
    for ligne in fichier:
        # Supprimer les espaces ou caractères de fin de ligne
        ligne = ligne.strip()
        
        # Ne retenir que les lignes contenant '::'
        if '::' in ligne:
            # Splitter la ligne sur '::'
            cle, valeur = ligne.split('::', 1)  # Utilisation de '1' pour ne splitter qu'une seule fois
            
            # Ajouter au dictionnaire
            dictionnaire_resultat[cle.strip()] = valeur.strip()

# Afficher le dictionnaire
print(dictionnaire_resultat)
```
